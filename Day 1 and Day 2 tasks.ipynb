# 3.Tell us difference between jupyter and notepad.

Jupyter is an interactive web application used for creating and sharing documents called notebooks, supporting multiple programming languages and rich media elements. Notepad, on the other hand, is a basic text editor for plain text editing with minimal features and no support for notebooks or advanced functionality.

# 4.Write Down difference between interpreter and compiler.

An interpreter executes code line by line, translating and executing each line immediately, making it slower but more interactive for debugging. A compiler translates the entire code into machine-readable instructions before execution, resulting in faster execution but less interactive debugging capabilities.

# 5.What is Data Structure, why we need one in programs?

A data structure is a way of organizing and storing data in a computer program. It provides a systematic way to store and manipulate data efficiently. Data structures are necessary in programs because they allow for efficient data retrieval, modification, and manipulation, enabling optimized algorithms and improved program performance.
